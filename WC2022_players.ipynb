```python
#Function description: It takes in two metrics in the form of y and x, and creates a scatter plot with a list of players
#specified and a dataframe. 
def plot_automator (y1,x1,players,stats,d):
    
    stats[y1] = stats[y1].astype(float)
    stats[x1] = stats[x1].astype(float)
    average_y = stats.get(y1).mean()
    average_x = stats.get(x1).mean()
    graph = stats.plot(kind='scatter', y=y1, x=x1)
    plt.title(y1 + ' vs ' + x1 + ' - WC 2022')
    
    for i in players:
        player = d[d.get('Player').str.contains(i)].index[0]
        plt.text(x=stats.get(x1)[player],y=stats.get(y1)[player],s=i,fontsize=10);
    
    #these will lines will retrieve players who are the highest on either metric but were not included in the list
    stats = stats.sort_values(by=y1, ascending=False)
    ind = stats.get(['Player']).index[0]
    name = stats['Player'].iloc[0]

    
    if name not in players:
        plt.text(x=stats.get(x1)[ind],y=stats.get(y1)[ind],s=name,fontsize=10);
    
    stats = stats.sort_values(by=x1, ascending=False)
    ind = stats.get(['Player']).index[0]
    name = stats['Player'].iloc[0]

    
    if name not in players:
        plt.text(x=stats.get(x1)[ind],y=stats.get(y1)[ind],s=name,fontsize=10);
    
    
    plt.axvline(x=average_x,alpha=1, color='red', label=x1,linewidth = 2)
    plt.axhline(y=average_y, alpha=1, color = 'red',label=y1, linewidth = 2);
plot_automator('Ast/90','xAG/90', ['Pedri','Bruyne','Neymar','Vinicius'],stats,f)
```

```python
def radar(player1, player2, stats_list,stats):
    
    #filter out the players not included
    new_stats = stats[(stats['Player'].str.contains(player1)) | (stats['Player'].str.contains(player2))]
    #new_stats = new_stats[stats_list]
    for i in stats_list:
        stats[i] = stats[i].astype(float)
        new_stats[i] = new_stats[i].astype(float)
   
    #get the ranges
    params = stats_list 
    ranges = []
    #a_values = []
    #b_values = []
    min_values = []
    max_values = []
    
    #min and max values for each metric
    for j in params: 
        a = min(stats[params][j])
        b = max(stats[params][j])
        
        ranges.append((a,b))
        
    #print(stats.columns)
    #for k in range(len(new_stats['Player'])):
   
    player1_df= stats[stats['Player'].str.contains(player1)]
    #print(type(player1_df))
    a_values = player1_df[stats_list].values.tolist()[0]
            
    player2_df= stats[stats['Player'].str.contains(player2)]
    b_values = player2_df[stats_list].values.tolist()[0]
    
    #remove the string name and keep the floating values 
    #a_values.remove(a_values[-1])
    #b_values.remove(b_values[-1])
    
    values = [a_values, b_values]
    
    #labels for the players
    title = dict(
        title_name = player1,
        title_color = 'red',
        subtitle_name = 'Player 1',
        subtitle_color = 'red',
        title_name_2 = player2,
        title_color_2 = 'blue',
        subtitle_name_2 = 'Player 2',
        subtitle_color_2 = 'blue',
        title_fontsize = 15,
        subtitle_fontsize = 12
    
    )
    
    endnote = "data via FBREF / Statsbomb"
    
    #instantiate the radar and put all the info in the plot
    radar = Radar()
    fig, ax = radar.plot_radar(ranges = ranges, params = params, values = values, radar_color=['red', 'blue'],
                              alphas = [.75,.6],title=title, endnote=endnote, compare=True)
    
```
